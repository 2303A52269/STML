```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


fig, axes = plt.subplots(3, 5, figsize=(20, 15))


mse_lasso, mae_lasso, rmse_lasso = np.array([]), np.array([]), np.array([])
mse_ridge, mae_ridge, rmse_ridge = np.array([]), np.array([]), np.array([])

models = {
    'KNN': (mse_knn, mae_knn, rmse_knn),
    'Linear Regression': (mse_lr, mae_lr, rmse_lr),
    'SVR': (mse_svr, mae_svr, rmse_svr),
    'Lasso': (mse_lasso, mae_lasso, rmse_lasso),
    'Ridge': (mse_ridge, mae_ridge, rmse_ridge),
}

metrics = ['MSE', 'MAE', 'RMSE']

for i, metric in enumerate(metrics):
    for j, (model_name, (mse, mae, rmse)) in enumerate(models.items()):
        data = mse if metric == 'MSE' else mae if metric == 'MAE' else rmse


        if data.size > 0:
            sns.histplot(data, ax=axes[i, j], kde=True)
            axes[i, j].set_title(f'Bootstrapped {metric} for {model_name}')
            axes[i, j].set_xlabel(metric)
            axes[i, j].set_ylabel('Frequency')
        else:
            axes[i, j].set_title(f'Bootstrapped {metric} for {model_name} (No Data)')

plt.tight_layout()
plt.show()
```
